```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #Adjusts the configuration of the plots we will create

#Read in the data

df = pd.read_csv(r'~/Downloads/movies.csv')
```

```python
# Let's look at the data
df.head()
```

```python
#Check if there's any data missing by loopoing through the columns
for col in df.columns:
    percent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percent_missing))
```

```python
# Column Data Types
df.dtypes
```

```python
# Changing the data type of the column
df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64') 
```

```python
df.head()
```

```python
#To create a correct year column
df['yearcorrect'] = df['released'].str.split('[(,\s)]{1}').str[3]

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False);
```

```python
pd.set_option('display.max_rows', None)
```

Dropping Any Duplicate Companies

```python
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Budget might have high correlation
# Company might have high correlation
```

Creating A Scatter Plot with Budget vs Gross Earnings

```python
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

Let's start looking at correlation

```python
df.corr(method='pearson', numeric_only=True)
#Only works on numerical fields
```

```python
correlation_matrix = df.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie features')
plt.show
```

```python
# Looks at Company
df.head()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized
```

```python
df.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson', numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie features')
plt.show
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
